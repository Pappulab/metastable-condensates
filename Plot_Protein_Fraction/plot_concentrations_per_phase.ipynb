# Plot amount of protein in each phase vs time

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

conc=[17.179869184, 21.47483648, 25, 26.8435456, 33.554432, 40, 41.9430, 50, 52.4288, 60, 65.536, 81.92, 100, 102.4, 128, 150, 160, 200]
conc2=[17, 21, 25, 26, 33, 40, 41, 50, 52, 60, 65, 81, 100, 102, 128, 150, 160, 200]

datatypeabove=['128_Day_4','200_Day_4','200_Day_4'] #degree of supersaturation: 0.59, 0.55, 0.50
#datatypeabove=['102_Day_4','160_Day_4','160_Day_4'] #degree of supersaturation: 0.36, 0.33, 0.28
#datatypeabove=['200_Day_4','200_Day_4','200_Day_4'] 
#datatypeabove=['128_Day_4','128_Day_4','128_Day_4'] 
mycolors=['#000000','#3954a5','#6cbd47']
mycolors2=['#b8b7b3','#8798d6','#477743']
construct=['WT','D262V','D262N']
csat = [71.2, 115.4, 121.2]

fig, axs = plt.subplots(1, 3, figsize=(15, 4))
#plt.figure(figsize=(3.5,3))

for c in range(0,len(construct)):
    
    tmp=datatypeabove[c].split('_')
    currconc=conc[conc2.index(int(tmp[0]))]
    currsupersat=str(round(math.log(currconc/csat[c]),2))

    
    
    # Get and plot experimental data
    dfexp=pd.read_table('../Code_Kiersten_Each_Condition_500_max_text_files/exp_fibril_'+construct[c]+'_'+datatypeabove[c]+'.txt', header=None)
    explist=dfexp[0].tolist()
    
    
    with open('../Code_Kiersten_Each_Condition_500_max_text_files/best_guess_'+construct[c]+'_'+datatypeabove[c]+'.txt') as myfile:
        for line in myfile:
            line = line.strip()
            line = line.split(",")
            fscal=(float(line[10]))
            
    explistscal=[]
    for i in explist:
        explistscal.append(float(i)/fscal)
    
    # Get and plot 
    
    
    axs[c].plot(explistscal,'-',color=mycolors[c], label='Exp Fibril')
    axs[c].set_ylabel('Protein Concentration (uM)')
    axs[c].set_xlabel('Time (min)')
    
    # Get and plot computational data - fibrils
    dfcompfibril=pd.read_table(construct[c]+'_'+datatypeabove[c]+'_3phase_protein_amount_Fibrils.txt', header=None)
    fibrillist=dfcompfibril[1].tolist()
    fibriltime=dfcompfibril[0].tolist()
    
    axs[c].plot(fibriltime,fibrillist,linestyle='--',color=mycolors2[c], label='Fit Fibril')
    
    # Get and plot computational data - Dilute phase
    dfcompdilute=pd.read_table(construct[c]+'_'+datatypeabove[c]+'_3phase_protein_amount_Dilute phase.txt', header=None)
    dilutelist=dfcompdilute[1].tolist()
    dilutetime=dfcompdilute[0].tolist()
    
    axs[c].plot(dilutetime,dilutelist,'-',color=mycolors2[c], label='Fit Dilute')
    
    # Get and plot computational data - Dense phase
    dfcompdense=pd.read_table(construct[c]+'_'+datatypeabove[c]+'_3phase_protein_amount_Dense phase.txt', header=None)
    denselist=dfcompdense[1].tolist()
    densetime=dfcompdense[0].tolist()
    
    axs[c].plot(densetime,denselist,linestyle='dotted',color=mycolors2[c], label='Fit Dense')
    
    axs[c].title.set_text(construct[c]+': '+str(currconc)+' uM, Sc = '+currsupersat)        
    axs[c].legend() 

#plt.xlim([0, 350])

plt.rcParams['pdf.fonttype'] = 42 # Makes text editiable
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.sans-serif'] = "Arial"
plt.rcParams['font.family'] = "sans-serif"
#plt.savefig('Protein_concentration_3phases_vs_time_sc_055_day_4.pdf', transparent=True, bbox_inches = "tight") 
#plt.savefig('Protein_concentration_3phases_vs_time_all_200_day_4.pdf', transparent=True, bbox_inches = "tight") 
#plt.savefig('Protein_concentration_3phases_vs_time_all_128_day_4.pdf', transparent=True, bbox_inches = "tight") 
```
